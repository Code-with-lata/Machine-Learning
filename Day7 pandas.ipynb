```python
# How to do slicing in pandas
import pandas as pd
var = pd.DataFrame({"A":[1,2,3,4,],"B":[9,8,7,6]})
var["ml"] = var["A"][:3]
var
```

```python
# insert any column at any dedicated index
import pandas as pd
var = pd.DataFrame({"A":[1,2,3,4,],"B":[9,8,7,6]})
var.insert(1,"Python",var["A"])
var
```

```python
# How to pop column from any dataframe
import pandas as pd
var = pd.DataFrame({"A":[1,2,3,4,],"B":[9,8,7,6]})
var2 = var.pop("B")
var
```

```python
# How to delete any column from any data frame
import pandas as pd
var = pd.DataFrame({"A":[1,2,3,4,],"B":[9,8,7,6]})
del var["A"]
var
```

```python
# how to merge a Dataframe
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4,],"B":[9,8,7,6]})
var2 = pd.DataFrame({"A":[1,2,3,4,],"C":[10,11,12,13]})
pd.merge(var1,var2,on="A")
```

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[9,8,7,6]})
var2 = pd.DataFrame({"A":[1,2,3,5],"C":[10,11,12,13]})#it remove automatacally mismatch data
pd.merge(var1,var2,on="A")
```

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[9,8,7,6]})
var2 = pd.DataFrame({"A":[1,2,3,5],"C":[10,11,12,13]})
pd.merge(var1,var2,how="left")
```

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[9,8,7,6]})
var2 = pd.DataFrame({"A":[1,2,3,5],"C":[10,11,12,13]})
pd.merge(var1,var2,how="right")
```

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[9,8,7,6]})
var2 = pd.DataFrame({"A":[1,2,3,5],"C":[10,11,12,13]})
pd.merge(var1,var2,how="outer")#outer work like a union
```

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[9,8,7,6]})
var2 = pd.DataFrame({"A":[1,2,3,5],"C":[10,11,12,13]})
pd.merge(var1,var2,how="outer",indicator=True)
```
