```python
# concatenate two data frames
import pandas as pd
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([d1,d2])
```

```python
# concatenate two data frames
import pandas as pd
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2],"B":[21,22]})
pd.concat([d1,d2],axis=1,join="outer")
```

```python
# key parameters in pandas
import pandas as pd
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})
pd.concat([d1,d2],axis=1,keys=["d1","d2"})
```

```python
# Group by in pandas
import pandas as pd
d1 = pd.DataFrame({"Name":["a","b","c","d","a","b","c"],"s1":[12,13,14,12,13,14,15],"s2":[22,23,24,22,23,24,25]})
d1
```

```python
# Group by in pandas
import pandas as pd
d1 = pd.DataFrame({"Name":["a","b","c","d","a","b","c"],"s1":[12,13,14,12,13,14,15],"s2":[22,23,24,22,23,24,25]})
d2= d1.groupby("Name")
for x,y in d2:
   print(x)
   print(y)
   
```

```python
# Melt function
import pandas as pd
d1 = pd.DataFrame({"days":[1,2,3,4,5,6],"eng":[10,12,14,15,16,12],"math":[17,18,19,13,14,16]})
pd.melt(d1)
```

```python
# pivot function is used to reshape
import pandas as pd
d1 = pd.DataFrame({"days":[1,2,3,4,5,6],"st_name":["a","a","a","a","a","a"],"eng":[10,12,14,15,16,12],"math":[17,18,19,13,14,16]})
d1.pivot(index="days",columns ="st_name")
```
