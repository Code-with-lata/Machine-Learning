```python
# which data in csv file is blank when we use thet data in pandas 
# then that particular data show NaN(not a number)
```

```python
#Remove row which has the NaN values 
csv_1.dropna(axis=0)
```

```python
##Remove column which has the NaN values 
csv_1.dropna(axis=1)
```

```python
##Remove row which has the NaN values 
csv_1.dropna(how="any")
```

```python
#remove all the row when wholw row data is NaN
csv_1.dropna(how="all")
```

```python
#how to fill the in place of NaN
csv_1.fillna("softpro")
```

```python
# it is use to fill the previous value in place of NaN
csv_1.fillna(method="ffill")#ffil (forword fill)
```

```python
#it is  use to fill the previous value in place of NaN
csv_1.fillna(method="bfill")#bfill(backword fill)
```

```python
#in this code it take the right value in place of NaN
csv_1.fillna(method="bfill",axis=1)
```

```python
#only put th value one time in a column
csv_1.fillna("python",limit=1)
```
