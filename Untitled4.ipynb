```python
import matplotlib.pyplot as plt
y = ["c","c++","java","python"]
x = [10,20,30,40]
explode = [0.1,0.0,0.0,0.0]
color = ["r","b","g","y"]
plt.pie(x,labels=y,radius=1.5,explode=explode,autopct="%0.1f%%",colors = color, shadow=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
y = ["c","c++","java","python"]
x = [10,20,30,40]

colors = ["r","b","g","y"]
plt.pie(x,labels=y,radius=1.5,colors = colors,)
plt.pie([1],colors="w")
plt.show()
```

```python
#Stack/Area plot

x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]
l = ["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels = l,colors=["r","g","b"])
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y= [2,3,1,4,5,3,6]
color = [10,20,30,40,50,60,70]
size = (500,300,600,300,500,700,800)
plt.xlabel("language",fontsize=15)
plt.ylabel("Days",fontsize=15)
plt.title("language",fontsize=20)
scatter =plt.scatter(x,y,c=color,marker ="*",s=size,cmap="viridis" ,edgecolor ="r")
colorbar = plt.colorbar(scatter)
colorbar.set_label("colorbar",fontsize=15)
plt.show()
```
