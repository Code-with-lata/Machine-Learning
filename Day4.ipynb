```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,6,8]
c = ["r","y","g","b"]
plt.xlabel("language",fontsize=15)
plt.ylabel("Day",fontsize=15)
plt.title("Softpro")
plt.bar(x,y,color=c)
plt.show
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,6,8]
c = ["r","y","g","b"]
sizes =[400,200,400,300]
plt.scatter(x,y,color=c,marker="*",s=sizes,edgecolor="g")
plt.xlabel("language",fontsize=15)
plt.ylabel("Day",fontsize=15)
plt.title("Softpro")
plt.show
```
